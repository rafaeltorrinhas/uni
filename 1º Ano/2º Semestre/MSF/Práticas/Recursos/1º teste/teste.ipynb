```python
t = np.array([0,48,96,144,192,240,288,336,384])
ativ = np.array([10.03,7.06,4.88,3.38,2.26,1.66,1.14,0.79,0.58])
N = len(t)
sum_xy = np.sum(t*ativ)
sum_y = np.sum(ativ)
sum_x = np.sum(t)
sum_xx = np.sum(np.square(t))
sum_yy = np.sum(np.square(ativ))

r2 = (N * sum_xy - sum_x * sum_y)**2 / \
    ((N * sum_xx - sum_x**2) * (N * sum_yy - sum_y**2))
m = ((N * sum_xy - sum_x * sum_y)/(N*(sum_xx)-((sum_x)**2)))
dm = np.absolute(m) * np.sqrt((1/r2)-1)/((N-2))
b = (((sum_xx * sum_y)-sum_x * sum_xy)/(N*sum_xx-((sum_x)**2)))

plt.xlabel("Tempo (h)")
plt.ylabel("Atividade (mBq)")
plt.plot(t, ativ,"o", color = "red")
plt.plot(t, m * t + b, "--")
plt.show()
print(f"rÂ² = {r2} \nm = {m}\nerro = {dm}")


t = np.array([0,48,96,144,192,240,288,336,384])
ativ = np.log(np.array([10.03,7.06,4.88,3.38,2.26,1.66,1.14,0.79,0.58]))
sum_xy = np.sum(t*ativ)
sum_y = np.sum(ativ)
sum_x = np.sum(t)
sum_xx = np.sum(np.square(t))
sum_yy = np.sum(np.square(ativ))



b = (((sum_xx * sum_y)-sum_x * sum_xy)/(N*sum_xx-((sum_x)**2)))
r2 = (N * sum_xy - sum_x * sum_y)**2 / \
    ((N * sum_xx - sum_x**2) * (N * sum_yy - sum_y**2))
m = ((N * sum_xy - sum_x * sum_y)/(N*(sum_xx)-((sum_x)**2)))
dm = np.absolute(m) * np.sqrt((1/r2)-1)/((N-2))

plt.plot(t, ativ, "o", color = "red")
plt.plot(t, m * t + b, "--")
plt.show()


print(f"rÂ² = {r2} \nm = {m}\nerro = {dm}")


t12 = -np.log(2)/m


print(f"\nt1/2 = {t12}")
```

```python

h0 = 2 
v0 = 15 
deg = 30 
vf = 20  
g = 9.8  

theta = np.radians(deg)

t_flight = (vf * np.sin(theta) + np.sqrt(vf**2 * np.sin(theta)**2 + 2 * g * h0)) / g

t = np.linspace(0, t_flight, 100)

x = 0 + v0 * np.cos(theta) * t + 0 * t ** 2

y = h0 + v0 * np.sin(theta) * t - 0.5 * g * t**2

plt.plot(x, y)
plt.xlabel("DistÃ¢ncia Horizontal (m)")
plt.ylabel("DistÃ¢ncia Vertical (m)")
plt.title("TrajetÃ³ria")
plt.grid(True)
plt.show()

```
