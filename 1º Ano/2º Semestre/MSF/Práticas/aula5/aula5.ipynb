```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([3,4])
b = np.array([4,-3])

plt.figure(figsize=(5,8))
plt.axis([-0.5,4.5,-3.5,4.5])
plt.arrow(0,0,a[0],a[1],color="k",width=0.1,length_includes_head=True)
plt.arrow(0,0,b[0],b[1],color="r",width=0.1,length_includes_head=True)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
